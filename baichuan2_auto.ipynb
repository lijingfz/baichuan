## 一级属性

```python
df = pd.read_csv('./res.csv')
df.head()
```

```python
titles = df['kouBeiTitle']
contents = df['content']
features = ['空间', '动力', '操控性', '经济性', '舒适性', '外观', '内饰', '科技智能', '新能源因素', '安全性', '配置', '品牌', '服务', '质量/可靠性', '价格/性价比']
feat_dict = {}
for feature in features:
    feat_dict[feature] = []
```

```python
def anylize(feature, text):
    messages = []
    messages.append({"role": "user", "content": f"以下是一段对于汽车的评论，评论哪些是明确涉及{feature}相关内容？仅回答与{feature}相关内容。 {text}"})
    response = model.chat(tokenizer, messages)
    print(response)
    messages.append({"role": "assistant", "content": response})
    
    messages.append({"role": "user", "content": f"根据上一步的回答及原文，评论是否有汽车的{feature}相关内容？ A是 B否， 仅回答A B"})
    response = model.chat(tokenizer, messages)
    print(response)

    if 'B' in response:
        return '未提及'

    messages.append({"role": "user", "content": f"根据上述内容，对该汽车车型{feature}属性进行评价进行总结。仅回答相关内容。 "})
    response = model.chat(tokenizer, messages)
    print(response)

    messages.append({"role": "assistant", "content": response})
    messages.append({"role": "user", "content": f"根据对{feature}属性的总结，评论是否对{feature}属性满意？ A满意 B不满意 C未提及 只回答A，B，C"})
    response = model.chat(tokenizer, messages)
    print(response)

    if 'C' in response:
        return '未提及'
    elif 'B' in response:
        return '不满意'
    elif 'A' in response:
        return '满意'
    else:
        return '未提及'
    
```

```python
%%time
for i, (title, content) in enumerate(zip(titles, contents)):
    text = title + '。'+ content
    try:
        print(i)
        for feature in features:
            print(feature)
            response = ''
            try:
                response = anylize(feature, text)
                print(f'{feature} : {response}')
                print('-------')
            except Exception as e:
                print(e)
            feat_dict[feature].append(response)
    except Exception as e:
        print('error:', e)

    brands.append(brandtext)
    types.append(typetext)
```

```python
for feature in features:
    df[feature] = feat_dict[feature]
```
